### Задание 1 (0.5 балла)

Используя `numpy` и `matplotlib`, начнём с тренировки: постройте график функций $sin(x)$ и $cos(x)$ на интервале от $-\pi$ до $\pi$, используя 100 точек. Добавьте легенду и сетку.. Используйте для линий разные цвета, одна из них пусть будет пунктирной. Результат в этом и следующих заданиях на построение графиков должен получиться примерно как на картинке.
````python
# ваш код здесь
x = np.linspace(-np.pi,np.pi, 100)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(5, 3))
plt.plot(x, sin, label='sin(x)', color='aqua')
plt.plot(x, cos, label='cos(x)', color='fuchsia',ls='--')
plt.title('sin и cos')
plt.xlabel('значение X')
plt.ylabel('значение функции')
plt.yticks(np.arange(-1, 1.1, 0.25))
plt.grid()
plt.legend()
plt.show()
````

### Задание 3 (1 балл)

Постройте subplots, где на первом графике будет круговая диаграмма, показывающая долю фильмов каждого рейтинга *(rating)*, а на втором столбчатая диаграмма, показывающая 3 страны, в которых было снято больше всего фильмов с рейтингом TV-MA.

*Если фильмов в каких-то рейтингах меньше 50, то их в график брать не нужно.*

+ Заголовки (можно жирным)

Круговая диаграмма:

Столбчатая диаграммма:
+ Сделайте так, чтобы ось Y была справа

````python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
ttt=sum(df_net.loc[df_net['type']=='Movie']['rating'].value_counts().to_frame().query('count>=50')['count'])
plt.pie(df_net.loc[df_net['type']=='Movie']['rating'].value_counts().to_frame().query('count>=50')['count'], labels=df_net.loc[df_net['type']=='Movie']['rating'].value_counts().to_frame().query('count>=50').index, autopct=lambda x: f'{int(x * ttt / 100)}',
        textprops={'fontfamily': 'monospace'})
plt.title('Доля фильмов по рейтингам', fontweight='bold', fontdict={'family': 'monospace'})
plt.axis('equal')

plt.subplot(1, 2, 2)
sns.barplot(x=df_net[(df_net['rating'] == 'TV-MA')&(df_net['type']=='Movie')].melt(id_vars=['title','director'], value_vars='country')['value'].value_counts().head(3).index,
            y=df_net[(df_net['rating'] == 'TV-MA')&(df_net['type']=='Movie')].melt(id_vars=['title','director'], value_vars='country')['value'].value_counts().head(3).values)
plt.title('Топ-3 стран', fontweight='bold', fontdict={'family': 'monospace'})
plt.ylabel('Колво', fontdict={'family': 'monospace'})
plt.xlabel(' ')
plt.gca().yaxis.set_label_position('right')
plt.gca().yaxis.tick_right()
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.show()
````

### Задание 8* (1 балл)

Постройте при помощи библиотеки plotly интерактивную столбчатую диаграмму, которая сравнивает среднюю продолжительность фильмов и сериалов за каждый год, начиная с 2000.

````python
# ваш код здесь
import plotly.express as px

df=df_net.loc[(df_net['year']>=2000)].groupby(['year','type'], as_index=False).agg({'min':'mean','season':'mean'})
fig = px.bar(df, x='year', y=['min','season'], color='type',
             barmode='group',
             title='Средняя продолжительность')

fig.update_layout(
    xaxis_title='Год',
    yaxis_title='Продолжительность'
)
fig.show()
````
